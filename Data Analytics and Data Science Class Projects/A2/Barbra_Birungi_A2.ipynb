```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
bnb_data = pd.read_csv('Airbnb_NYC_2019.csv')
bnb_data
```

```python
bnb_data['neighbourhood_group'].apply(lambda name: 'Manhattan' in name).sum()
```

```python
manhattan_listing = bnb_data.loc[['Manhattan' in name for name in bnb_data['neighbourhood_group']],:]
manhattan_listing
```

```python
manhattan_listing.shape
```

```python
plt.hist(manhattan_listing['price'], bins = np.arange(0,1000,40))
plt.title('Manhattan Air_bnb price listing')
plt.xlabel('price')
plt.ylabel('count')
plt.show()
```

```python
plt.scatter(x=manhattan_listing['price'], y=manhattan_listing['number_of_reviews'], s=5)
plt.xlim(0,1100)
plt.title('Number of reviews vs price for the Manhattan Listing')
plt.xlabel('Price in $')
plt.ylabel('Number of reviews')
plt.show()
```

```python
top_ten_listing = manhattan_listing.groupby(['neighbourhood']).size().sort_values(ascending=False).reset_index(name = 'counts').head(10)
top_ten_listing
```

```python
top_ten_listing.shape
```

```python
top_ten = top_ten_listing['neighbourhood'].to_list()
top_ten
```

```python

top_neighnbourhood = manhattan_listing[manhattan_listing['neighbourhood'].isin(top_ten)]
top_neighnbourhood
```

```python
top_neighnbourhood.shape
```

```python
plt.hist(manhattan_listing['price'], bins= np.arange(0,1100,40) )
plt.title('The Price distribution for the Manhattan listing')
plt.xlabel('price in USD')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize =(12, 6))
sn.scatterplot(x='price', y='number_of_reviews', data= manhattan_listing, s=10)
plt.xlim(0,2500)
plt.title('Number of Reviews over Price for the Manhattan listing')
plt.xlabel('Price in USD')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
sn.countplot(x='neighbourhood', data= top_neighnbourhood)
plt.xticks(rotation= 45, horizontalalignment='right')
plt.title('Plot showing the count of the Top-ten Manhattan neighbourhoods')
plt.xlabel('Top-ten neighbourhoods')
plt.ylabel('count')
plt.show()
```

```python

sn.barplot(x= 'neighbourhood', y='counts', data= top_ten_listing, ci=False)
plt.xticks(rotation= 45, horizontalalignment='right')
plt.show()
```
