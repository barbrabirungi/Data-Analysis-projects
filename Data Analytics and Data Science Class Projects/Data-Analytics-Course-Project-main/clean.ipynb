```python
#create dataframes for the different types of entries. 
# I will then calculate the sum of all views. 
# I would note that because the data comes from several different sources, 
# the sum will likely be double counting some of the views
def get_language(Page):
    res = re.search('[a-z][a-z].wikipedia.org',Page)
    if res:
        return res[0][0:2]
    return 'na'

Wiki_data['lang'] = Wiki_data.Page.map(get_language)

from collections import Counter

print(Counter(Wiki_data.lang))
lang_sets = {}
lang_sets['en'] = Wiki_data[Wiki_data.lang=='en'].iloc[:,0:-1]
lang_sets['ja'] = Wiki_data[Wiki_data.lang=='ja'].iloc[:,0:-1]
lang_sets['de'] = Wiki_data[Wiki_data.lang=='de'].iloc[:,0:-1]
lang_sets['na'] = Wiki_data[Wiki_data.lang=='na'].iloc[:,0:-1]
lang_sets['fr'] = Wiki_data[Wiki_data.lang=='fr'].iloc[:,0:-1]
lang_sets['zh'] = Wiki_data[Wiki_data.lang=='zh'].iloc[:,0:-1]
lang_sets['ru'] = Wiki_data[Wiki_data.lang=='ru'].iloc[:,0:-1]
lang_sets['es'] = Wiki_data[Wiki_data.lang=='es'].iloc[:,0:-1]

sums = {}
for key in lang_sets:
    sums[key] = lang_sets[key].iloc[:,1:].sum(axis=0) / lang_sets[key].shape[0]

#to show how the number of views changes over time, i plot
days = [r for r in range(sums['en'].shape[0])]

fig = plt.figure(1,figsize=[10,10])
plt.ylabel('Views per Page')
plt.xlabel('Day')
plt.title('Pages in Different Languages')
labels={'en':'English','ja':'Japanese','de':'German',
        'na':'Media','fr':'French','zh':'Chinese',
        'ru':'Russian','es':'Spanish'
       }

for key in sums:
    plt.plot(days,sums[key],label = labels[key] )
    
plt.legend()
plt.show()
 
```

```python
#QUESTION 7
import re

def get_access(Page):
    r = re.search("mobile-web",Page)
    x = re.search("all-access", Page)
    y = re.search("desktop", Page)

    if r:
        return r[0]
    elif x:
        return x[0]
    elif y:
        return y[0]

Wiki_data_days['access'] = Wiki_data_days.Page.apply(get_access)


from collections import Counter
print(Counter(Wiki_data_days.access))

Wiki_data_days.head()

#Counter({'all-access': 342210, 'mobile-web': 107604, 'desktop': 99186})
# all-access is used most followed by desktop then mobile-web
```
